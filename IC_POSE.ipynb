```python
import os
import glob

# Percorsi statici
video_folder = "/content/estratti_video"
output_base = "/content/output_video"
mmdet_cfg = "demo/mmdetection_cfg/rtmdet_m_640-8xb32_coco-person.py"
mmdet_ckpt = "checkpoints/rtmdet.pth"
pose_cfg = "configs/wholebody_2d_keypoint/rtmpose/coco-wholebody/rtmpose-x_8xb32-270e_coco-wholebody-384x288.py"
pose_ckpt = "checkpoints/rtmpose-x.pth"
device = "cpu"  # oppure "cuda" se usi GPU

# Trova tutti i file .mp4 in tutte le sottocartelle
video_files = glob.glob(os.path.join(video_folder, '**', '*.mp4'), recursive=True)

# Crea cartella output principale
os.makedirs(output_base, exist_ok=True)

# Esegui elaborazione per ogni video trovato
for video_path in video_files:
    # Estrai percorso relativo rispetto a video_folder
    rel_path = os.path.relpath(video_path, video_folder)
    base_name = os.path.splitext(rel_path.replace(os.sep, "_"))[0]  # es: Test_12_13_25_0
    output_dir = os.path.join(output_base, base_name)
    os.makedirs(output_dir, exist_ok=True)

    print(f"🔄 Elaborazione di: {video_path} -> {output_dir}")

    # Comando di esecuzione
    !python demo/topdown_demo_with_mmdet.py \
        {mmdet_cfg} \
        {mmdet_ckpt} \
        {pose_cfg} \
        {pose_ckpt} \
        --input "{video_path}" \
        --output-root "{output_dir}" \
        --save-predictions \
        --device {device}

print("✅ Tutti i video sono stati elaborati.")
```

```python
import json
import numpy as np
import cv2
import os

# Carica il file JSON
with open('/content/mmpose/vis_results/results_input.json', 'r') as f:
    data = json.load(f)

instances_per_frame = data['instance_info']
kp_colors = np.array(data['meta_info']['keypoint_colors']['__ndarray__'])
skeleton = data['meta_info']['skeleton_links']
skeleton_colors = np.array(data['meta_info']['skeleton_link_colors']['__ndarray__'])

# Crea directory per i frame PNG con trasparenza
transparent_dir = '/content/mmpose/vis_results/'
os.makedirs(transparent_dir, exist_ok=True)

canvas_h, canvas_w = 720, 1280

for frame_idx, frame_data in enumerate(instances_per_frame):

    # TRASPARENTE
    img = np.zeros((canvas_h, canvas_w, 4), dtype=np.uint8)  # RGBA con alpha = 0 (trasparente)

    # BIANCO
    #img = np.full((canvas_h, canvas_w, 4), fill_value=255, dtype=np.uint8)  # Bianco (R=255, G=255, B=255, A=255)

    instances = frame_data['instances']
    for inst in instances:
        keypoints = np.array(inst['keypoints'])  # (133, 2)
        for i, (x, y) in enumerate(keypoints):
            if x > 0 and y > 0:
                color = tuple(map(int, kp_colors[i])) + (255,)  # RGBA
                cv2.circle(img, (int(x), int(y)), 2, color, -1, lineType=cv2.LINE_AA)

        for link_idx, (i, j) in enumerate(skeleton):
            xi, yi = keypoints[i]
            xj, yj = keypoints[j]
            if xi > 0 and yi > 0 and xj > 0 and yj > 0:
                scolor = tuple(map(int, skeleton_colors[link_idx])) + (255,)  # RGBA
                cv2.line(img, (int(xi), int(yi)), (int(xj), int(yj)), scolor, 1, lineType=cv2.LINE_AA)

    frame_path = os.path.join(transparent_dir, f"frame_{frame_idx:04d}.png")
    cv2.imwrite(frame_path, img)
```
